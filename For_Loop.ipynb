```python
lower = int(input("enter the lower range:"))
upper = int(input("enter the upper range:"))

for i in range(lower, upper+1):
  for j in range(1,11):
    print(i*j, end=" ")
  print()
```

```python
lower = int(input("enter the lower range:"))
upper = int(input("enter the upper range:"))

for i in range(lower, upper+1):
  for j in range(1,6):
    print(str(i) + "x" + str(j) + "=" + str(i*j), end=" ")
  print()
```
