```python
#program for rock paper scissors
from random import randint

print("Welcome to Rock, Paper, Scissors!")
print("Select from below items")
player = input("Rock, Paper, Scissors: ")
player = player.lower()
print("your choice is", player)

#for computer choices
randnum = randint(1,3)
if randnum == 1:
  computer = "rock"
elif randnum == 2:
  computer = "paper"
elif randnum == 3:
  computer = "scissors"
print("computer choice is", computer)

if player == computer:
  print("tie")
elif player == "rock":
  if computer == "paper":
    print("you lose")
  elif computer == "scissors":
    print("you win")
elif player == "paper":
  if computer == "rock":
    print("you win")
  elif computer == "scissors":
    print("you lose")
else:
  print("invalid input")
```
